```python
import os
dir_path = 'C:/Users/heenatigalakanat3478/peterlucas/withpits/'

file_names = os.listdir(dir_path)

# Filter out image files (e.g., by extensions)
image_paths = [(os.path.join(dir_path, name),name) for name in file_names if name.lower().endswith(('.png', '.jpg', '.jpeg'))]
print(image_paths)
```

```python
for img,name in image_paths:
    img=image.load_img(img,target_size=(1280,1280))
    img=image.img_to_array(img)
    img=np.expand_dims(img,axis=0)
    img=preprocess_input(img)
    result=model.predict(img)
    
    results.append([name,result])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
test_data_dir = 'C:/Users/heenatigalakanat3478/peterlucas/'  # Update this path
img_height, img_width = 1280,1280  # Same as during training
batch_size = 16  # Can adjust based on your system's memory

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode=None,  # Change this if needed
    shuffle=False)  # Important for mapping filenames

print(test_generator.filenames)
# Make predictions
predictions = model.predict(test_generator, steps=np.ceil(test_generator.samples/test_generator.batch_size))

# Map predictions to labels
predicted_labels = (predictions > 0.5).astype(int)

# Compile results
results = list(zip(test_generator.filenames, predicted_labels))

# Print or save the results
for filename, label in results:
    print(f"Filename: {filename}, Predicted Label: {label[0]}")

import csv
with open('C:/Users/heenatigalakanat3478/Saskatchewan Polytechnic/Peter Lucas Gravel Pit AI_GRP - General/ARD Grant/Models/Models for google Images/Restnet152V2 testanalysisnew1.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(results)
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import os
from PIL import Image

# Load the pre-trained model
model_path = 'C:/Users/heenatigalakanat3478/models/ResNet152V2.h5'
model = load_model(model_path)

# Specify the folder containing images to predict
image_folder = 'C:/Users/heenatigalakanat3478/peterlucas/withnonpits/'
img_height, img_width = 1280, 1280 # These should match the input shape the model expects

# Initialize the ImageDataGenerator for normalization (as done during training)
datagen = ImageDataGenerator(rescale=1./255)

# Function to load and prepare the image
def load_image(img_path, show=False):
    img = Image.open(img_path)
    img = img.resize((img_width, img_height))
    img_array = np.array(img)
    img_array = img_array / 255.0  # Normalize the image
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

with open('C:/Users/heenatigalakanat3478/Saskatchewan Polytechnic/Peter Lucas Gravel Pit AI_GRP - General/ARD Grant/Models/Models for google Images/Restnet152V2 testanalysisnew2-2withnonpits.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Image Name', 'Predicted Class', 'Raw Prediction'])
# Loop through all images in the folder, make predictions, and print them
    for img_name in os.listdir(image_folder):
        img_path = os.path.join(image_folder, img_name)
        img = load_image(img_path)
        prediction = model.predict(img)
        predicted_class = int(prediction > 0.5) # Assuming 0.5 as the threshold for binary classification
        print(f"Image: {img_name} | Predicted class: {'Non-pit' if predicted_class == 0 else 'Pit'} | Raw prediction: {prediction}")
        writer.writerow([img_name, 'Non-pit' if predicted_class == 0 else 'Pit', float(prediction)])





# Note: The threshold of 0.5 used here is a common default for binary classification. Depending on your application, you might want to adjust this threshold to tune the sensitivity/specificity of your model predictions.
```
